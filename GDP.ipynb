```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('countries of the world.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
for column in data.columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

### EDA

```python
df=data.drop(['Country','Region'],axis=1)
```

```python
df.head()
```

```python
#replacin "," with "."
df = df.replace(',','.', regex=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df1 = data[['Country','Region']]
```

```python
df1.head()
```

```python
df = df.astype(float)
```

```python
df3 = pd.concat([df1,df],axis=1,join='inner')
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3.columns
```

### Country vs population 

```python
temp1 = df3.sort_values('Population',ascending = False)[:50]
plt.figure(figsize=(26,8))
sns.barplot(x ='Country',y = 'Population', data = temp1,palette='Paired')
plt.xticks(rotation = 'vertical')
plt.show()
```

### Country vs Area

```python
temp2 = df3.sort_values('Area (sq. mi.)',ascending = False)[:50]
plt.figure(figsize=(26,8))
sns.barplot(x ='Country',y = 'Area (sq. mi.)', data = temp2)
plt.xticks(rotation = 'vertical')
plt.show()
```

### Country vs infant mortality rate


```python
temp3 = df3.sort_values('Infant mortality (per 1000 births)',ascending = False)[:50]
plt.figure(figsize=(26,8))
sns.barplot(x ='Country',y = 'Infant mortality (per 1000 births)', data = temp3,palette='Paired')
plt.xticks(rotation = 'vertical')
plt.show()
```

### GDP

```python
temp4 = df3.sort_values('GDP ($ per capita)',ascending = True)[:50]
plt.figure(figsize=(26,8))
sns.barplot(x ='Country',y = 'GDP ($ per capita)', data = temp4)
plt.xticks(rotation = 'vertical')
plt.show()
```

### GDP vs Infant mortality

```python
temp5 = df3.sort_values('GDP ($ per capita)',ascending = True)[:100]
plt.figure(figsize=(26,8))
sns.barplot(x='GDP ($ per capita)',y='Infant mortality (per 1000 births)' ,data = temp5)
plt.show()
```

### Litteracy vs GDP

```python
temp6 = df3.sort_values('GDP ($ per capita)',ascending = True)[:50]
plt.figure(figsize=(26,8))
sns.barplot(x='GDP ($ per capita)',y='Literacy (%)' ,data = temp6)
plt.show()
```

### Birthrate vs deathrate

```python
plt.figure(figsize=(28,8))
temp7 = df3.sort_values('GDP ($ per capita)',ascending = True)[:50]
plt.style.use('ggplot')
sns.lineplot(x='Birthrate',y='Deathrate',data=temp7)
plt.xticks(rotation=0);
```

### Usage of phones in countries

```python
plt.figure(figsize=(28,8))
temp8 = df3.sort_values('Phones (per 1000)',ascending = True)[:50]
plt.style.use('ggplot')
sns.lineplot(x='Country',y='Phones (per 1000)',data=temp8)
plt.xticks(rotation=90);
```

### usage of phone vs GDP

```python
plt.figure(figsize=(28,8))
temp9 = df3.sort_values('Phones (per 1000)',ascending = False)[:50]
sns.barplot(x='GDP ($ per capita)',y='Phones (per 1000)',data=temp9)
plt.xticks(rotation=90);
```

### Region vs netmigration

```python
plt.figure(figsize=(28,8))
temp10 = df3[:50]
sns.boxplot(x='Region',y='Net migration',data=temp10)
plt.xticks(rotation=90);
```
